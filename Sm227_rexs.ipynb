# Sm2Ir2O7 REXS

```python
from core.lattice import *
from matplotlib import pyplot as plt
%matplotlib inline
```

## Scattering tensors

```python
Sm227 = Lattice([10.3],'F d -3 m')
```

```python
# calculate ATS tensor from absorbing site and scattering vector
r = [0,0,0]
q = [10,0,0]

Fats = Sm227.generate_tensor(r,q)
```

```python
# calculate resonant magnetic tensor from magnetic structure factor
Fm   = Sm227.xrms_tensor([1,0,0])
```

## Diffraction setup

```python
energy       =  11217
polarization = [1,0]
surface      = [1,0,0]
aziref       = [0,1,1]

Sm227.diffraction(energy, polarization, surface, aziref)
```

```python
df = Sm227.diffract.azimuthal_scan(Fats + Fm, q, absorb=False, plot=True)
```
